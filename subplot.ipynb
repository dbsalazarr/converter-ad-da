### Generando más de un gráfico

```python
import matplotlib.pyplot as plt 
import numpy as np
import math
```

```python
# Set de datos
x = np.linspace(0, 21, 12)
y = x**2
```

```python
# Mostrando la gráfica y un histograma
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')

plt.subplot(1, 2, 2)
plt.hist(y)
plt.show()
```

```python
# Mostrando la gráfica y un histograma
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.plot(y, x, 'b*--')
plt.subplot(1, 2, 2)
plt.pie(y)
plt.show()
```

```python
# Mostrando la gráfica y un histograma
plt.subplot(2, 1, 1)
plt.plot(x, y, 'r--')
plt.plot(y, x, 'b*--')
plt.subplot(2, 1, 2)
plt.pie(y)
plt.show()
```
